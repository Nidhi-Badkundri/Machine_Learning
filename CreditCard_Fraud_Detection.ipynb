# Introduction

Credit card fraud detection is a crucial aspect of financial security and risk management in the banking and financial industry.
Detecting and preventing credit card fraud is essential to protect both financial institutions and their customers from financial losses and maintain trust in electronic payment systems.

The dataset for credit card fraud detection typically consists of a collection of transactions, each labeled as either legitimate or fraudulent.

The goal is to create robust models capable of distinguishing between normal and fraudulent activities, thereby enabling real-time or near-real-time intervention when suspicious transactions occur.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings as w
w.filterwarnings("ignore")

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

## 1. Data Collection

>Id -   A unique Id for each row.

>Time -   Number of seconds elapsed between this transaction and the first transaction in the dataset

>V1-V28 -   Features after dimensionality reduction to protect user identities and sensitive features

>Amount -   Transaction amount

>Class -   Target Class (1 for fraudulent transactions, 0 genuine)

```python
import imblearn
from imblearn.under_sampling import RandomUnderSampler
from collections import Counter
```

```python
us = RandomUnderSampler(random_state=1,replacement=True)
x_us, y_us = us.fit_resample(x,y)
```

```python
print("orignal: ", Counter(y))
print("reshaped: ",Counter(y_us))
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
lr = LogisticRegression()
lr.fit(x_us,y_us)

ypred_train=lr.predict(xtrain)
ypred_test= lr.predict(xtest)

print("Training data : ")
print(classification_report(ytrain,ypred_train))

print("Testing data : ")
print(classification_report(ytest,ypred_test))
```

```python
lr = LogisticRegression(solver = 'liblinear')
lr.fit(x_us,y_us)

ypred_train=lr.predict(xtrain)
ypred_test= lr.predict(xtest)

print("Training data : ")
print(classification_report(ytrain,ypred_train))

print("Testing data : ")
print(classification_report(ytest,ypred_test))
```

```python
#Scaling to use solver - sag or saga
sc = StandardScaler()
xtrain = sc.fit_transform(xtrain)
xtest = sc.transform(xtest)
```

```python
lr = LogisticRegression(solver = 'sag')
lr.fit(x_us,y_us)

ypred_train=lr.predict(xtrain)
ypred_test= lr.predict(xtest)

print("Training data : ")
print(classification_report(ytrain,ypred_train))

print("Testing data : ")
print(classification_report(ytest,ypred_test))
```

```python
lr = LogisticRegression(solver = 'saga')
lr.fit(x_us,y_us)

ypred_train=lr.predict(xtrain)
ypred_test= lr.predict(xtest)

print("Training data : ")
print(classification_report(ytrain,ypred_train))

print("Testing data : ")
print(classification_report(ytest,ypred_test))
```

```python
from sklearn.metrics import accuracy_score

print("Training data")
print(accuracy_score(ytrain,ypred_train))

print("\n Testing data")
print(accuracy_score(ytest,ypred_test))
```

```python
from sklearn.metrics import confusion_matrix

print("Training data")
print(confusion_matrix(ytrain,ypred_train))

print("\n Testing data")
print(confusion_matrix(ytest,ypred_test))
```

```python
from sklearn.metrics import recall_score,precision_score
```

```python
print(recall_score(ytrain,ypred_train))
```

```python
print(precision_score(ytrain,ypred_train))
```

```python
#Checking probability train and test
ytrain_prob=lr.predict_proba(xtrain)
ytrain_prob
```

```python
ytrain_prob=lr.predict_proba(xtrain)[:,1]
ytrain_prob
```

```python
ytest_prob=lr.predict_proba(xtest)
ytest_prob
```

```python
from sklearn.metrics import precision_recall_curve, roc_auc_score
```

```python
p,r,th=precision_recall_curve(ytrain,ytrain_prob)
```

```python
#calculate AUC-PRC using roc_auc_score
auc_prc = roc_auc_score(ytrain,ytrain_prob)
auc_prc
```

```python
#plot PRC curve

plt.step(p,r,th,color='b',alpha=0.2,where='post')
plt.fill_between(p,r, step='post',alpha=0.2,color='b')

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.ylim([0.0,1.05])
plt.xlim([0.0,1.0])
plt.title(f'Precision-Recall Curve (AUC-PRC={auc_prc: 2f})')
plt.show()


print(f'AUC-PRC: {auc_prc:.4f}')
```
